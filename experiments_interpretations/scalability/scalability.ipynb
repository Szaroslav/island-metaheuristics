```python
import json
from statistics import mean
import matplotlib.pyplot as plt
```

```python
data_40 = json.load(open("data/n40.json", "r"))
print(data_40)
```

```python
data_100 = json.load(open("data/n100.json", "r"))
print(data_100)
```

```python
data_200 = json.load(open("data/n200.json", "r"))
print(data_200)
```

```python
data_40_ips = [value['ips'] for value in data_40.values()]
data_100_ips = [value['ips'] for value in data_100.values()]
data_200_ips = [value['ips'] for value in data_200.values()]
```

```python
plot_data_40 = (list(range(len(data_40_ips))), data_40_ips)
plot_data_100 = (list(range(len(data_100_ips))), data_100_ips)
plot_data_200 = (list(range(len(data_200_ips))), data_200_ips)
```

```python
plt.plot(plot_data_40[0], plot_data_40[1])
plt.title("Ilość iteracji algorytmu na sekundę dla każdej wyspy.\n Uruchomienie dla 40 wysp.")
plt.xlabel("numer wyspy")
plt.ylabel("ilość iteracji na sekundę")
plt.show()
```

```python
plt.plot(plot_data_100[0], plot_data_100[1])
plt.title("Ilość iteracji algorytmu na sekundę dla każdej wyspy.\n Uruchomienie dla 100 wysp.")
plt.xlabel("numer wyspy")
plt.ylabel("ilość iteracji na sekundę")
plt.show()
```

```python
plt.plot(plot_data_200[0], plot_data_200[1])
plt.title("Ilość iteracji algorytmu na sekundę dla każdej wyspy.\n Uruchomienie dla 200 wysp.")
plt.xlabel("numer wyspy")
plt.ylabel("ilość iteracji na sekundę")
plt.show()
```

```python
mean_data_40_ips = mean(data_40_ips)
mean_data_100_ips = mean(data_100_ips)
mean_data_200_ips = mean(data_200_ips)
```

```python
plt.plot([40, 100, 200], [mean_data_40_ips, mean_data_100_ips, mean_data_200_ips])
plt.title("średnia ilość iteracji na sekundę dla wysp w zależności od\n ilości wysp uruchomionych naraz.")
plt.xlabel("ilość wysp")
plt.ylabel("średnia ilość iteracji na sekundę")
plt.show()
```
